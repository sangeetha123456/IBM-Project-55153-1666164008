**REPLACING THE OUTLIERS**

```python
ot = []
dt = []
at = df["Age"]
mt = df["IsActiveMember"]
for i in at:
  if (30 <= at[i] <= 40):
    ot.append(at[i])
    dt.append(mt[i])
    

x = dt
y = ot

fig = px.scatter(x =x, y=y)
fig.show()

```
